In this section, we will examine the fundamental algorithms used in quantum computers. These algorithms were developed in the early stages of quantum computation and are considered the first algorithms to enhance computational speed in quantum computers. We refer to these algorithms as "The Canon" because they have become standard in the field of quantum computation.

Many of these algorithms were discussed at a high level in previous sections. However, in this section, we are examining them in more detail. Although some of these algorithms require still-developing quantum computers, analyzing them now can provide us with a better understanding of future possibilities.

Some of the algorithms discussed in this section are known as "black-box" or "oracle model" quantum algorithms. These types of algorithms allow us to determine how an unknown underlying function operates. By querying this function, we can learn how certain inputs are associated with certain outputs. This is a powerful tool for solving specific problems in quantum computers.

...now let's provide a bit more information about each of these.

- **GO (Grover Operator)**: Grover's search algorithm provides an efficient way to find a desired item in a quantum computer. It is faster than classical algorithms and offers a quadratic speedup.

- **QFT (Quantum Fourier Transform)**: Quantum Fourier Transform is the quantum version of the classical Fourier transform. It is particularly used in fields like number theory and data compression.

- **HHL (Harrow, Hassidim, Lloyd)**: The HHL algorithm is a quantum algorithm designed to solve linear equation systems by leveraging the advantages of quantum computers. It can provide advantages over classical computers, especially in solving large linear equation systems.

- **VQE (Variational Quantum Eigensolver)**: VQE is a quantum algorithm used to compute the electronic structures of chemical compounds and molecules. It can deliver faster results compared to classical computers and is an important tool in the analysis of molecular structures.

- **SIM (Simulation Matching Simulation)**: This concept involves comparing hardware quantum results with simulator results. Results obtained from real quantum devices are evaluated by comparing them with simulation results of the same algorithm. This is used to test the performance and accuracy of the hardware.

Before delving into the discussion of algorithms, there are some fundamental concepts we need to understand. One of these is the Query Model and its significance.

- *The Query Model is an approach used to determine the performance of quantum algorithms. In this model, both a quantum computer and a classical computer are given access to the same oracle (function), and the effectiveness of both computers in performing a specific task using the oracle is compared.*

- *For example, some quantum algorithms such as Deutsch's algorithm and Bernstein-Vazirani algorithm enable fast computations with the oracle. This allows proving lower bounds or exact expressions for the number of queries needed in both classical and quantum cases.*

- *More importantly, in some cases, "real" quantum speedups can be achieved with effective implementation of the oracle. For instance, Shor's quantum factoring algorithm provides a real speedup by using a polynomial number of gates that scale with the size of the input register of the oracle. Such algorithms constitute fundamental evidence demonstrating that quantum computers can indeed be faster than classical computers for certain tasks. Therefore, the query model is an important tool for understanding and demonstrating the potential advantages of quantum computers.*
